# Missing Value
1. Getting rid by - dropna()
2. filling them up - fillna()
   fillna() - A. Forward fill
              B. Backward fill
              

```python
df["Age"].fillna(df["Age"].mean())
```

```python
df["New Salary"].fillna(df["New Salary"].median())
```

```python
df["Age"].fillna(method = "ffill") # forward fill
```

```python
df["Age"].fillna(method = "bfill") # backward fill
```

```python
df
```

```python
df["Name"] = df["Name"].replace("Charlie", "Rose")
```

```python
# Duplicate Value

df_dup = df[df.duplicated(keep = "last")]
df_dup
```

```python
# invalid values
# Lambda ->

df["New Salary"] = df["New Salary"].apply(lambda x: x/10 if x > 65000 else x)
```
